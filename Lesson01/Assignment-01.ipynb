````python
def create_grammar(grammar_str, split = '=', line_split = '\n', word_split = '|'):
    grammar = {}
    for line in grammar_str.split(line_split): # 分成一行一行的
        if not line.strip(): continue # line为空则继续
        exp, con_exp = line.split(split)  # exp和con_exp分开
        grammar[exp.strip()] = [i.split() for i in con_exp.split(word_split)] # 存储字典
    return grammar
````

````python
create_grammar(animal, split = '=', line_split = '\n', word_split = '|')
````

````python
import random
````

````python
def generate(gram, target):
    if target not in gram: return target #目标不在grammar中，表示不需要深入
    res = [generate(gram, i) for i in random.choice(gram[target])] # 随机产生，对每个词判断是否需要深入
    return ''.join(res)
````

````python
example_gram = create_grammar(animal, split = '=', line_split = '\n', word_split = '|')
````

````python
generate(gram = example_gram, target = "animal")
````

````python
def generate_n(n):
    sentences = []
    random_choice = random.choice(choice)
    for i in range(n):
        sentence = generate(create_grammar(random_choice[1], split = '=', line_split = '\n', word_split = '|'), random_choice[0])
        sentences.append(sentence)
    return sentences
````

````python
import re
````

````python
def token(string):
    return re.findall('\w+',string)
````

````python
comments_clean = [''.join(token(str(com))) for com in comments]
````

````python
comments_clean
````

````python
import jieba #分词
````

````python
def cut(string):
    return list(jieba.cut(string))
````

````python
TOKEN = []
count = 0
````

````python
for i in comments_clean:
    if count % 50 == 0:
        print(count)
    if count > 10000:
        break
    count += 1
    TOKEN += cut(i)
````

````python
from collections import Counter
````

````python
words_count = Counter(TOKEN)
````

````python
# 试试可视化
````

````python
frequencies = [b for a,b in words_count.most_common(100)]
x = [x for x in range(100)]
import matplotlib.pyplot as plt
plt.plot(x, frequencies)
````

````python
def pro_1(word): # 单个词概率
    return words_count[word] / len(TOKEN)
````

````python
pro_1('电影')
````

````python
pro_1('爱国')
````

````python
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-1]))]
````

````python
TOKEN_2_GRAM
````

````python
words_count_2 = Counter(token_2_gram)
````

````python
words_count_2
````

````python
def pro_2(word_1, word_2): #两个词组合概率
    if word_1 + word_2 in token_2_gram:
        return words_count_2[word_1 + word_2] / len(TOKEN_2_GRAM)
    else:
        return 1 / len(TOKEN_2_GRAM)
````

````python
pro_2('看', '电影')
````

````python
pro_2('太', '恐怖')
````

````python
def get_sentence_probablity(sentence): #判断句子是否合理
    words = cut(sentence)
    sentence_pro = 1
    for index, word in enumerate(words[:-1]):
        next_word = sentence[index + 1]
        sentence_pro *= pro_2(word, next_word)
    return sentence_pro
````

>

````python
def generate_best(gram, model, n): # you code here
    sentences_pro_list = []
    for i in range(n):
        sentence = generate(create_grammar(gram[0]), gram[1])
        sentence_pro = model(sentence)
        sentence_pro_list = (sentence, sentence_pro)
        sentences_pro_list.append(sentence_pro_list)
    print(sentences_pro_list)
    return sorted(sentences_pro_list,key=lambda x:x[1], reverse=True)[0]
````
