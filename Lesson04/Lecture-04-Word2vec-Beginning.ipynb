## Step-02: Using https://github.com/attardi/wikiextractor to extract the wikipedia corpus

第二步：使用python wikipedia extractor抽取维基百科的内容（这里用的是gensim工具包中的WikiCorpus）

```python
import jieba #分词
from langconv import * # 将繁体转换为简体
from gensim.corpora import WikiCorpus  #将wikipedia内容抽取出来

def cut(string):
    """jieba分词"""
    return list(jieba.cut(string))

def simplified(string):
    """繁体=>简体"""
    return Converter('zh-hans').convert(string)

def extract_wikipedia_corpus(filename):
    i = 0
    with open("./wiki.txt", "w", encoding="utf-8") as f:
        wiki = WikiCorpus(filename, lemmatize=False, dictionary={})
        for text in wiki.get_texts():
            res = []
            for sentence in text:
                simplified_sentence = simplified(sentence)
                seg_with_jieba = cut(simplified_sentence)
                for word in seg_with_jieba:
                    res.append(word)
            f.write(' '.join(res) + '\n')
            i += 1
            
            if i % 50 == 0:
                print("Saved {} articles".format(str(i)))
                
        f.close()
        
if __name__ == "__main__":
    zhwiki_file_name = "./zhwiki-20190720-pages-articles-multistream.xml.bz2"
    extract_wikipedia_corpus(zhwiki_file_name)
```

第三步：参考Gensim的文档和Kaggle的参考文档，获得词向量。 注意，你要使用Jieba分词把维基百科的内容切分成一个一个单词，然后存进新的文件中。然后，你需要用Gensim的LineSentence这个类进行文件的读取。

在训练成词向量Model.

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence #文件读取
import multiprocessing #多进程
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(lineno)d -  %(message)s')

def word2vec_train():
    """词向量训练"""
    f_wiki = open("./wiki.txt", "r", encoding="utf-8")
    sentences = LineSentence(f_wiki)
    model = Word2Vec(sentences, size = 200, window = 10, min_count = 500, workers = multiprocessing.cpu_count()) # 可视化快一些
    model.save("./wiki.zh.model")
    model.wv.save_word2vec_format("./wiki.zh.vectors", binary = False)
```

```python
word2vec_train()
```

## Step-04: Using some words to test your preformance. 

第四步，测试同义词，找几个单词。

```python
word2vec_model = Word2Vec.load("./wiki.zh.model")
#查看词向量
print("江西", word2vec_model['江西']) # 获得单个单词的词向量
print(word2vec_model.similarity("江西","福建")) # 比较两个单词的相似度
sim_words = word2vec_model.most_similar('江西') # 获得近义词
for w in sim_words:
    print(w)
```

## Step-05: Using visualization tools: https://www.kaggle.com/jeffd23/visualizing-word-vectors-with-t-sne

第五步：使用Kaggle给出的T-SEN进行词向量的可视化。

```python
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(lineno)d -  %(message)s')

def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    print("冒一个泡")
    new_values = tsne_model.fit_transform(tokens[:5000]) # 电脑不行，不要逞强 = =

    x = []
    y = []
    print("冒两个泡")
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    print("冒三个泡")
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```
